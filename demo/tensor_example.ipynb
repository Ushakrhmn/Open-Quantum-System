```python
import quimb as qu
import matplotlib.pyplot as plt
import quimb.tensor as qtn
import random
```

```python
N = 64
```

```python
qc = qtn.Circuit(N)
```

```python
# just applly a H gate on one of the qubits

for _ in range(32):
    q1 = random.randint(0, N - 1)
    qc.apply_gate("H", q1)

for _ in range(10):
    q1 = random.randint(0, N - 1)
    q2 = random.randint(0, N - 1)
    while q1 == q2:
        q2 = random.randint(0, N - 1)
    qc.apply_gate("CNOT", q1, q2)
```

```python
samples = []
for s in  qc.sample(512, seed = 42):
    samples.append(s)
```

```python
print(samples)
```

```python
plt.hist(samples)
```

```python
V = qc.uni.draw()
```
